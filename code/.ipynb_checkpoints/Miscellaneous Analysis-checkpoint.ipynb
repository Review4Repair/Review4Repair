```python
import json
def getJsonData(JsonFile):
    with open(JsonFile, encoding="utf8") as f:
        data = json.load(f)
    return data
```

# Ziphs Distribution

```python
hard_freq_code = getJsonData('jsons/sorted_code_freq_100.json')
soft_freq_code = getJsonData('jsons/sorted_code_freq_soft.json')
hard_freq_comment = getJsonData('jsons/sorted_comment_freq_100.json')
soft_freq_comment = getJsonData('jsons/sorted_comment_freq_soft.json')
```

```python
hard_freq_code_list = []
for key in hard_freq_code:
    hard_freq_code_list.append(hard_freq_code[key])
    
soft_freq_code_list = []
for key in soft_freq_code:
    soft_freq_code_list.append(soft_freq_code[key])
    
hard_freq_comment_list = []
for key in hard_freq_comment:
    hard_freq_comment_list.append(hard_freq_comment[key])
    
soft_freq_comment_list = []
for key in soft_freq_comment:
    soft_freq_comment_list.append(soft_freq_comment[key])
```

## Data Distribution

```python
train = getJsonData('jsons/train_100.json')
test = getJsonData('jsons/test_100.json')
```

```python
type(test)
```

```python
dataset = train + test
```

```python
len(dataset)
```

```python
project_count = {}
for x in dataset:
    if x['project_name'] not in project_count:
        project_count[x['project_name']] = 0
    project_count[x['project_name']] += 1
```
